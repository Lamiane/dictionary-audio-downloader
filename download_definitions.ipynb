```python
import os
import re
import argparse
import traceback
import urllib.request
# import wget
```

```python
dictionary_url = "https://www.dictionary.com/browse/"

word_pre = "css-1jzk4d9 e1rg2mtf8\">"
word_suf = "</h"
audio_pre = "<audio preload"
pos_pre = "class=\"luna-pos\">"
pos_suf = "</span>"
section_end = "</section>"
trans_pre = "<span class=\"one-click-content css-1p89gle e1q3nk1v4\">"
trans_suf = "</span>"

pattern_word = re.compile(f"{word_pre}(.*?){word_suf}+.*?{audio_pre}")
pattern_section = re.compile(pos_pre+".*?"+section_end)
pattern_pos = re.compile(pos_pre+"(.*?)"+pos_suf)
definition_example_pattern = re.compile(f"{trans_pre}(.*?{trans_suf})")
pattern_definition = re.compile("(.*?)span")
pattern_example = re.compile("luna-example italic\">(.*?)</span>")

link = re.compile("<a .*?>(.*?)</a>")
```

```python
def get_all_translations(html_content, hide=True):
    word = pattern_word.findall(html_content)[0]
    print(word, '\n')
    
    posums = []
    for section in pattern_section.findall(html_content):
        section = re.sub(link, r'\1', section)
        
        pos = pattern_pos.findall(section)[0]
        defs = definition_example_pattern.findall(section)
        
        if defs:
            print('\n', pos, '\n')
            dupums = []
            for d in defs:
                dd = pattern_definition.findall(d)[0].strip('/').strip('<').strip()
                ex = pattern_example.findall(d)
                if len(dd) == 0:
                    continue
                print(dd)
                if ex:
                    ex = ex[0]
                    if hide:
                        ex = re.sub(word+"(.*?) ", r'***\1 ', ex)
                    print(ex,'\n')
                    dupums.append((dd, ex))
                else:
                    dupums.append((dd,))
            posums.append((pos, dupums))
            
    return posums
```

```python
with open("examplary_word_list", "r") as f:
    all_words = f.readlines()
all_words = [w.strip().lower() for w in all_words]

not_in_dictionary = []
for word in all_words:
    # get the dictionary.com HTML
    try:
        # retreive HTML
        responose = urllib.request.urlopen(dictionary_url+word)
        html_content = responose.read().decode('utf-8')
    except:
        traceback.print_exc()
        not_in_dictionary.append(word)
        continue

    all_translations = get_all_translations(html_content)
    print('\n\n-------------\n\n\n')
```
